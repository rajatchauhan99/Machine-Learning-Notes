# Row sampling

Creating a function for row sampling

```python
def row_sampling(df, percent):
    return df.sample(int(percent*len(df)), replace = True)
```

Making three sample datasets using row sampling

```python
df1 = row_sampling(df, 0.2)
df1
```

```python
df2 = row_sampling(df, 0.2)
df2
```

```python
df3 = row_sampling(df, 0.2)
df3
```

Let us train three decision trees using these three sampled datasets and see how well they do the predictions

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt1 = DecisionTreeClassifier()
dt2 = DecisionTreeClassifier()
dt3 = DecisionTreeClassifier()
```

```python
dt1.fit(df1.iloc[:,0:5],df1.iloc[:,-1])
dt2.fit(df2.iloc[:,0:5],df2.iloc[:,-1])
dt3.fit(df3.iloc[:,0:5],df3.iloc[:,-1])
```

Let us plot these trees

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(dt1)
```

```python
plot_tree(dt2)
```

```python
plot_tree(dt3)
```

# Column Sampling

Creating a function for column sampling

```python
df.columns
```

```python
len(df.columns)
```

```python
def column_sampling(df, percent):
    df_features = df.iloc[:,0:5]
    df_new = df_features.sample(int(percent*(len(df.columns)-1)), axis = 1)
    df_new["Target"] = df["Target"]
    return df_new
```

```python
df1 = column_sampling(df,0.8)
df1
```

```python
df2 = column_sampling(df,0.8)
df2
```

```python
df3 = column_sampling(df,0.8)
df3
```

Again training decision tress using these datasets now

```python
dt1 = DecisionTreeClassifier()
dt2 = DecisionTreeClassifier()
dt3 = DecisionTreeClassifier()

dt1.fit(df1.iloc[:,0:4],df1.iloc[:,-1])
dt2.fit(df2.iloc[:,0:4],df2.iloc[:,-1])
dt3.fit(df3.iloc[:,0:4],df3.iloc[:,-1])
```

Let us see the plots

```python
plot_tree(dt1)
```

```python
plot_tree(dt2)
```

```python
plot_tree(dt3)
```

We did column sampling, we will need to give data point for prediction in same order as the features selected.

# Combined Row and Column sampling

Creating a function for this

```python
def combined_sampling(df, percentage_row, percentage_column):
    df_row = df.sample(int(percentage_row*len(df)))
    df_features = df_row.iloc[:,0:5]
    df_new = df_features.sample(int(percentage_column*(len(df.columns)-1)), axis = 1)
    df_new["Target"] = df_row["Target"]
    return df_new   
```

Selecting 20 rows and 4 columns sampling

```python
df1 = combined_sampling(df, 0.2, 0.8)
df1
```

```python
df2 = combined_sampling(df, 0.2, 0.8)
df2
```

```python
df3 = combined_sampling(df, 0.2, 0.8)
df3
```

Now, we can train and see the plots

```python
dt1 = DecisionTreeClassifier()
dt2 = DecisionTreeClassifier()
dt3 = DecisionTreeClassifier()

dt1.fit(df1.iloc[:,0:4],df1.iloc[:,-1])
dt2.fit(df2.iloc[:,0:4],df2.iloc[:,-1])
dt3.fit(df3.iloc[:,0:4],df3.iloc[:,-1])
```

```python
plot_tree(dt1)
```

```python
plot_tree(dt2)
```

```python
plot_tree(dt3)
```
