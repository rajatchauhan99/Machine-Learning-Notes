# Working with the IRIS dataset

```python
from sklearn.datasets import load_iris
```

Creating an object of this dataset

```python
iris = load_iris()
iris
```

Defining features variable

```python
X = iris.data
X
```

```python
print(len(X))
```

Defining output variable

```python
y = iris.target
y
```

Dividing data into train and test using sklearn

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

# Importing Decision Tree Sklearn library

```python
from sklearn.tree import DecisionTreeClassifier
```

Creating an object of this Decision Tree Classifier

```python
clf = DecisionTreeClassifier()
```

Let us use training data now to train the model

```python
clf.fit(X_train, y_train)
```

Let us see the predictions of this model

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
y_test
```

Wow mostly it looks correct only

Let us see compare the predictions with the actual data and see the accuracy scores

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

So algorithm working quite well

Let us try to see things visually now

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(clf)
```

Wow we can see the actual tree how algorithm have defined nodes, branches and leaf

We will use Matplotlib code to increase the figure size

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(80,60))
plot_tree(clf)
```

So we can see untill we get gini impurity as zero then the algorithm stops and we get the leaf node.

And the leaf node will tell the occurance of how many times the type of flower or what the output we are getting with how many samples we are getting in this leaf.

# Working with some other dataset

Downloading Social_Network_Ads dataset from Kaggle

https://www.kaggle.com/datasets/rakeshrau/social-network-ads/ 

```python
import numpy as np
import pandas as pd
```

Creating a dataframe

```python
df = pd.read_csv('/Users/rajatchauhan/Desktop/Machine Learning Notes/Datasets/Social_Network_Ads.csv')
df
```

But here one of the feature 'Gender' is having categorical values

So we will convert it back to 0 and 1 using replace function.

```python
df['Gender'].replace({'Male': 0, 'Female' : 1}, inplace = True)
df
```

Now setting up input variable X, we do not require the User ID column as it is not a feature,

we will just use features as Gender, Age and EstimatedSalary based on that we will see if they purchased or not

```python
X = df.iloc[:,1:-1]
X
```

Setting up output variable which is Purchased or not

```python
y = df.iloc[:,-1]
y
```

```python
y.shape
```

Let us directly create the decision tree classifier object now

```python
clf1 = DecisionTreeClassifier()
```

Now we will not be going to use train test split as we are not interested on that, we will directly go for training the model to the entire dataset

```python
clf1.fit(X,y)
```

```python
plt.figure(figsize=(80,60))

plot_tree(clf1)
```

This is a very huge decision tree and a very deep decision tree is not advised as it can lead to overfitting problem

That is why it is important to control the depth of tree which is one of the hyperparameter as input we can give to the decision tree model.

This we can discuss in depth later.

Let us try to set the max_depth to some number and see our decision tree model plot

```python
clf2 = DecisionTreeClassifier(max_depth=3)
```

```python
clf2.fit(X, y)
```

Now let us look at the tree how does it look

```python
plt.figure(figsize=(80,60))

plot_tree(clf2)
```

This looks much better and more generalised model (Avoids overfitting issue)

This way here, we will predict the output based on the majority class in the leaf node.

As that would be having more probability.
