BLOB stands for *Binary Large OBject*. It is a data type that can store binary data. BLOB is generally a collection of large complex data in the Database. The make_blobs() function sometimes can be used to generate blobs of points with a Gaussian distribution.

## Working with Perfectly Linear Dataset

```python
from sklearn.datasets import make_blobs
```

Generate isotropic Gaussian blobs for clustering.

Parameters
----------
n_samples : number of samples (data points) of the dataset.

n_features : The number of features for each sample. X => [x1, x2...]

centers : The number of classes or centers of data, means the possible discreate values of output variable y. (0,1,2 and so on) or Clusters of data

cluster_std : The standard deviation of the clusters.

random_state : 
    Determines random number generation for dataset creation. Pass an int
    for reproducible output across multiple function calls.

Returns
-------
X : ndarray of shape (n_samples, n_features)
    The generated samples.

y : ndarray of shape (n_samples,)
    The integer labels for cluster membership of each sample.

We will create a sample clustering data with 2 features and 2 clusters

```python
X, y = make_blobs(n_samples=50,n_features = 2, centers=2,
                  random_state=0, cluster_std=0.60)
```

# Creating a function to create a best fit line

```python
def plot_svc_decision_function(model, ax=None, plot_support=True):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    # linspace returns evenly spaced numbers over a specified interval.
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    
    # plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0],
                   model.support_vectors_[:, 1],
                   s=300, linewidth=1, facecolors='none');
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

```python
plt.scatter(x1, x2, c=y, s=50, cmap='winter')
plot_svc_decision_function(model);
```

## The importance of Support Vectors

```python
def plot_svm(N=10, ax=None):
    X, y = make_blobs(n_samples=200, centers=2,
                      random_state=0, cluster_std=0.60)
    X = X[:N]
    y = y[:N]
    model = SVC(kernel='linear', C=1E10)
    model.fit(X, y)
    
    ax = ax or plt.gca()
    ax.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='winter')
    ax.set_xlim(-1, 4)
    ax.set_ylim(-1, 6)
    plot_svc_decision_function(model, ax)

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)
for axi, N in zip(ax, [60, 120]):
    plot_svm(N, axi)
    axi.set_title('N = {0}'.format(N))
```

## Working with Almost Linearly Separable Dataset

```python
X, y = make_blobs(n_samples=100, centers=2,
                  random_state=0, cluster_std=1.2)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='winter');
```

```python
X, y = make_blobs(n_samples=100, centers=2,
                  random_state=0, cluster_std=0.8)

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)

for axi, C in zip(ax, [100.0, 0.01]):
    model = SVC(kernel='linear', C=C).fit(X, y)
    axi.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='winter')
    plot_svc_decision_function(model, axi)
    axi.scatter(model.support_vectors_[:, 0],
                model.support_vectors_[:, 1],
                s=300, lw=1, facecolors='none');
    axi.set_title('C = {0:.1f}'.format(C), size=14)
```
